```python
#Student Performance analytics System
import pandas as pd

# Step 1: Load Dataset
df = pd.read_csv(r"C:\Users\bibek\Downloads\student_performance_project.csv")

print("Original Dataset")
print(df.head())
```

```python
print(df.shape)
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
# Fill missing marks with column mean
df["Math"] = df["Math"].fillna(df["Math"].mean())
df["Science"] = df["Science"].fillna(df["Science"].mean())

# Fill missing City with Unknown
df["City"] = df["City"].fillna("Unknown")
```

```python
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()

print(df.shape)
```

```python
print(df.duplicated().sum())
```

```python
top_students = df[df["Math"] > 80]
print(top_students)
```

```python
print(df["Math"].mean())
print(df["Science"].mean())
print(df["English"].mean())

print(df["Math"].max())
print(df["Math"].min())
```

```python
# Total Marks
df["Total"] = df["Math"] + df["Science"] + df["English"]

# Average Marks
df["Average"] = df["Total"] / 3

# Ranking (Highest Average = Rank 1)
df["Rank"] = df["Average"].rank(method="dense", ascending=False)

print(df[["Name","Total","Average","Rank"]])
```

```python
df["Result"] = df.apply(
    lambda x: "Pass"
    if x["Math"] >= 35 and x["Science"] >= 35 and x["English"] >= 35
    else "Fail",
    axis=1
)

print(df[["Name","Math","Science","English","Result"]])
```

```python
print(df["Result"].value_counts())
```

```python
print(df.groupby("Gender")[["Math","Science","English"]].mean().round(2))

print(df.groupby("City")[["Math","Science","English"]].mean().round(2))
```

```python
report = df.groupby(["Gender","City"])[["Math","Science","English","Total","Average"]].mean().round(2)

print(report)
```

```python
topper = df.sort_values("Average", ascending=False)

print(topper.head(5))
```

```python
df.to_csv("student_performance_final_report.csv", index=False)

print("Final Report Exported Successfully!")
```
